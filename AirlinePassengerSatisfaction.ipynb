```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Reading Dataset

```python
data = pd.read_csv("airline-dataset.csv")
data.head()
```

```python
data.shape
```

```python
data.describe()
```

###### The collection contains roughly 100K passenger entry records.

```python
data.info()
```

###### Since the first two columns won't be useful for factor analysis, we may remove them.

```python
data = data.drop(data.iloc[:,[0, 1]], axis = 1)
```

```python
cat_indexes = [0, 1, 3, 4] + list(range(6, 20))
data.iloc[:,cat_indexes] = data.iloc[:,cat_indexes].astype('category')
```

```python
data.isnull().sum()
```

###### Arrival Delay in Minutes contains about 310 Null values.

```python
#there are null values in the 'Arrival Delay in Minutes'
# meadian is used to fill those 
data['Arrival Delay in Minutes'].fillna(data['Arrival Delay in Minutes'].median(axis = 0), inplace = True)
#check the null and get the sum
data.isnull().sum()
```

###### The column's median was used to fill in the missing data. Dataset is currently NULL-free.

## One Hot Encoding

```python
# Import label encoder
from sklearn import preprocessing

# The object label_encoder is capable of comprehending word labels.
lbEncoder = preprocessing.LabelEncoder()

# Encode labels in column 'species'.
data['Gender']= lbEncoder.fit_transform(data['Gender'])
data['Customer Type']= lbEncoder.fit_transform(data['Customer Type'])
data['Type of Travel']= lbEncoder.fit_transform(data['Type of Travel'])
data['Class']= lbEncoder.fit_transform(data['Class'])
data['satisfaction']= lbEncoder.fit_transform(data['satisfaction'])
```

# Split Train Test

```python
#import model selection from sklearn
from sklearn.model_selection import train_test_split

#drop columns
X = data.drop(columns=['Gender','Departure Delay in Minutes','Arrival Delay in Minutes','satisfaction','Gate location'])
#output variable
y = data['satisfaction']
  
# using the train test split function
X_train, X_test, y_train, y_test = train_test_split(X,y ,random_state=3, test_size=0.2 )

X.head(5)
```

## Decision Tree Classifier

```python
#import metrics from sklearn
from sklearn import metrics
#import decision tree from sklearn
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)
y_pred1 = tree.predict(X_test)
accuracy1 = metrics.accuracy_score(y_test, y_pred1)

#The precision score is a useful measure of the success of prediction when the classes are very imbalanced.
precision1 = metrics.precision_score(y_test, y_pred1)

print("accuracy score :",accuracy1)
print("precision score:",precision1)
```

## Logistic Regression

```python
#import ensemble form sklearn
from sklearn.linear_model import LogisticRegression

regression = LogisticRegression()
regression.fit(X_train, y_train)
y_pred2 = regression.predict(X_test)
accuracy2 = metrics.accuracy_score(y_test, y_pred2)

#The precision score is a useful measure of the success of prediction when the classes are very imbalanced.
precision2 = metrics.precision_score(y_test, y_pred2)
print("accuracy:",accuracy2)
print("precision:",precision2)
```

## Random Forest

```python
#import ensemble form sklearn 
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier()
forest.fit(X_train, y_train)
y_pred3 = forest.predict(X_test)
accuracy3 = metrics.accuracy_score(y_test, y_pred3)

#The precision score is a useful measure of the success of prediction when the classes are very imbalanced.
precision3 = metrics.precision_score(y_test, y_pred3)
print("accuracy score :",accuracy3)
print("precision score:",precision3)
```

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train, y_train)
y_pred4 = svc.predict(X_test)
accuracy4 = metrics.accuracy_score(y_test, y_pred4)

#The precision score is a useful measure of the success of prediction when the classes are very imbalanced.
precision4 = metrics.precision_score(y_test, y_pred4)
print("accuracy:",accuracy4)
print("precision:",precision4)
```
